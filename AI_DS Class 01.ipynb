### String

```python
name = "city"
print(name.upper())
print(name.lower())
print(name.count("c")) #count no of time c appears
print(name.startswith("c")) #to check if string starts with c or not
print(name.endswith("a")) #to check if string ends with c or not
print(len(name))
```

#### String Formatting

```python
name = input("Enter your name: ")
age = input("Enter your age: ")
# to discuss
print("My name is "+ name +" and I am "+str(age)+" yrs old")
print(f"My name is {name} and I am {age} yrs old")  #f string
print("My name is {0} and my age is {1}",format(name,age)) #string Formatting
```
